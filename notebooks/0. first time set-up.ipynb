```python
from dbx_utils import ingest_setup

dbutils.widgets.text("table_catalog", "") # Name of catalog for the endpoints table
dbutils.widgets.text("table_schema", "") # Name of schema for the endpoints table
dbutils.widgets.text("table_name", "") # Name of table to store endpoints for download
dbutils.widgets.text("table_managed_location", "") #set managed location (abfss://container@storage.dfs.core.windows.net/)


dbutils.widgets.text("volume_catalog", "") # Name of catalog for the endpoints table
dbutils.widgets.text("volume_schema", "") # Name of schema for the endpoints table
dbutils.widgets.text("volume_name", "") # Name of table to store endpoints for download
```

```python
# Create catalog, schema and endpoints table. If catalog and schema already exist, only endpoint table is created

catalog = dbutils.widgets.get("table_catalog")
schema = dbutils.widgets.get("table_schema")
table = dbutils.widgets.get("table_name")
managed_location = dbutils.widgets.get("table_managed_location")

ingest_setup.create_endpoint_table(catalog=catalog, schema=schema, table=table, managed_location=managed_location)
```

```python
# Create volume for the temporary storage of raw downloads

catalog = dbutils.widgets.get("volume_catalog")
schema = dbutils.widgets.get("volume_schema")
volume_name = dbutils.widgets.get("volume_name")

ingest_setup.create_volume(catalog=catalog, schema=schema, volume=volume_name)
```
